```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
dataset=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
dataset['label']=breast_cancer.target
dataset.head()
```

```python
X=dataset.drop(columns='label',axis=1)
Y=dataset['label']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),# Input Layer
    keras.layers.Dense(20,activation='relu'),# Hidden Layer
    keras.layers.Dense(2,activation='sigmoid')# Output Layer
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
```

```python
history=model.fit(X_train_scaled,Y_train,epochs=10,validation_split=0.1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test)
```

```python
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```
